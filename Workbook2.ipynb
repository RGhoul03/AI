# Задание 1.3.1

Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и
единицами.

```python
import numpy as np

values = np.zeros((8, 8))

for i in range(0, 8):
    if i % 2 == 0:
         for j in range(0, 7, 2):
              values[i][j] = 1
    else:
         for j in range(1, 8, 2):
              values[i][j] = 1

print(values)
```

# Задание 1.3.2

Создать 5x5 матрицу со значениями в строках от 0 до 4. Для создания
необходимо использовать функцию arange.

```python
import numpy as np

arr = np.arange(25).reshape(5, 5)

for i in range(0, 5):
    arr[i] = np.arange(0, 5)

print(arr)
```

# Задание 2.3.1

Найти евлидово расстояние  между двумя Series(точками) a и b, не используя встроенную формулу

```python
import pandas as pd
import math as m

a = pd.Series([0, 1, 1], ['x', 'y', 'z'])
b = pd.Series([1, 1, 0], ['x', 'y', 'z'])

res = m.sqrt((a['x']-b['x'])**2 + (a['y']-b['y'])**2 + (a['z']-b['z'])**2)

print(res)
```

# Задание 3.3

Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака 
(sepal_length_cm) с использованием минимаксного преобразования, а 
второго (sepal_width_cm) с задействованием z-масштабирования.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

minMax = MinMaxScaler()
std = StandardScaler()

url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
arr = pd.read_csv(url)
display(arr.head(5))

arr[['sepal_length_cm']] = minMax.fit_transform(arr[['sepal_length_cm']])
arr[['sepal_width_cm']] = std.fit_transform(arr[['sepal_length_cm']])

arr.head(5)
```
